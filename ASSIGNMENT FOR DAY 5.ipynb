```python
## 3. plot a Pie Chart of Indian Car companies market share % values
# :. max ( 6 Companies market share )

from matplotlib import pyplot as plt
import numpy as np
  
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES','volkwagon','toyota','honda','general motors']
  
data = [23, 17, 35, 29, 12, 41,100,55,33,66]
  
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)
plt.show()
```

```python
# 4. Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )


states = ['kerala','maharastra','karnataka','delhi','tamilnadu','up']
  
data = [14000,5200,4400,6003,5000,2000]
plt.bar(states,data)
plt.show()
```
